#### Investigate Non-Numeric Measurement Values

```python
df.rename(columns={"TERM APRIL 2010": "THERM APRIL 2010"}, inplace=True)

string_columns = df.select_dtypes(include="string").columns

string_columns
```

```python
energy_columns = [
    column
    for column in string_columns
    if "KWH" in column or "THERM" in column
]

energy_columns
```

```python
deferred_conversion_columns = [
    "ELECTRICITY ACCOUNTS",
    "GAS ACCOUNTS",
]
```

```python
validated_numeric_columns = [
    column
    for column in string_columns
    if column not in energy_columns
       and column not in deferred_conversion_columns
       and "COMMUNITY" not in column
       and "BUILDING" not in column
]

validated_numeric_columns
```

```python
numeric_candidate_columns = (
        energy_columns + validated_numeric_columns
)

numeric_candidate_columns
```

```python
invalid_numeric_values = {}

for column in numeric_candidate_columns:
    cleaned_values = (
        df[column]
        .dropna()
        .str.replace(",", "", regex=False)
    )

    invalid_values = cleaned_values[
        pd.to_numeric(cleaned_values, errors="coerce").isna()
    ].unique()

    if len(invalid_values) > 0:
        invalid_numeric_values[column] = invalid_values

if invalid_numeric_values:
    print("Columns containing unexpected non-numeric values:\n")

    for column, values in invalid_numeric_values.items():
        print(f"{column}: {values}\n")
else:
    print("Validation successful: "
          "All candidate numeric columns contain valid numeric values."
          )
```

#### Non-Numeric Measurement Values Summary

Validation identified the value "Less than 4" in the **ELECTRICITY ACCOUNTS** and **GAS ACCOUNTS** variables. These columns otherwise contain numeric values and represent quantitative account counts.

The available dataset documentation confirms that both variables represent utility accounts counts but does not explain the meaning or intended treatment of the value "Less than 4". Because its exact numeric value cannot be determined from the available documentation, no transformation will be applied to these columns at this stage.

These variables have been separated into a dedicated list of deferred conversion columns and will not be converted during this phase of the cleaning process. This decision preserves the integrity of the original data and avoids introducing unsupported assumptions into the analysis. As a result, these variables remain stored as text for the remainder of the cleaning process. This reflects a temporary implementation decision rather than a change in their intended analytical meaning as quantitative variables.

#### Convert Validated Numeric Columns

The candidate numeric columns were validated before conversion to ensure that they contained only numeric values (after removing formatting characters such as thousands separators). Two variables (`ELECTRICITY ACCOUNTS` and `GAS ACCOUNTS`) contained the undocumented value **"Less than 4"** and were excluded from conversion pending additional authoritative guidance.

All remaining validated numeric columns are converted to numeric data types in the following step using `errors="raise"` to ensure that any unexpected values are reported immediately.


```python
for column in numeric_candidate_columns:
    df[column] = pd.to_numeric(
        df[column].str.replace(",", "", regex=False),
        errors="raise"
    )
```

#### Investigate Identifier Variables

The `CENSUS BLOCK` variable is an identifier rather than a quantitative measurement. Although pandas imported the column as `float64`, this was caused by the presence of a missing value, which required the use of a floating-point data type.

The online dataset stores census block values without decimal places, confirming that the decimal portion introduced by pandas is a storage artifact rather than part of the original data. To better reflect the semantic meaning of the variable, `CENSUS BLOCK` will be converted to a string identifier while preserving missing values.

```python
df["CENSUS BLOCK"] = (
    df["CENSUS BLOCK"]
    .astype("Int64")
    .astype("string")
)
```

```python
categorical_text_columns = [
    "COMMUNITY AREA NAME",
    "BUILDING TYPE",
    "BUILDING_SUBTYPE",
]

for column in categorical_text_columns:
    df[column] = df[column].astype("string")
```
